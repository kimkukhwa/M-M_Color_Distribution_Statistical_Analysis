# Part 1. Web-scraping
### This web scraping project is to extract the 48 Individual M&M Color Distribution Data datasets from Josh Madison's websites (Thank you very much for the M&M experiment!). Data source: https://joshmadison.com/2007/12/02/mms-color-distribution-analysis-individual-pack-data/

*   Two methods, BeautifulSoup & Pandas read_htlm, are used
*   Interested datasets are in table format  


```python
# Setup
import pandas as pd
import requests
import json
from bs4 import BeautifulSoup

pd.options.mode.chained_assignment = None 
URL = "https://joshmadison.com/2007/12/02/mms-color-distribution-analysis-individual-pack-data"
```

## Method 1 BeautifulSoup

```python
#BeautifulSoup test

res = requests.get(URL)
soup = BeautifulSoup(res.text, 'html.parser')
tables = soup.find_all('table')
table = tables[1]

# Print the html table - but it is not meaningful yet
'print(table)
```

```python
# Print the first table - but it is still not pretty. It requires some clearning.
df = pd.read_html(str(table), header=0,flavor='bs4')[0]
print(df.head(15).to_string())
```

## Method 2: pd.read_html

## Clean the Tables

```python
# Set column names and remove rows & colums
for _,table in enumerate(table_MN):
  table.columns=['Color','Blue',	'Brown',	'Green',	'Orange',	'Red',	'Yellow','todelete']
  table.drop('todelete',
          axis=1,
          inplace = True
          )
  table.drop([5],
             axis=0,
             inplace=True
             ) 

 
 
  
```

```python
# Transpose tables for better view
t_tables=[]
for _,table in enumerate(table_MN):
  table=table.T
  table.reset_index(inplace=True)
  table.columns=table.iloc[0]
  table=table[1:]
  t_tables.append(table)

```

```python
t_tables[0]
```

## Concat all Tables to One Data Frame

```python
frames = [ table for table in t_tables ]
package_n= [i for i in range(1,49)]
mnm_df = pd.concat(frames,keys=package_n)
```

```python
mnm_df
```

## More Cleaning!

```python
mnm_df.reset_index(level=[0,1],inplace=True)
mnm_df
```

```python
mnm_df.rename(columns={'level_0': 'package_n'}, inplace=True)
mnm_df.drop('level_1',axis=1,inplace=True)
```

```python
mnm_df.info()
```

```python
mnm_df['Qty observed'] = mnm_df['Qty observed'].astype(int)
mnm_df['Qty expected'] = mnm_df['Qty expected'].astype(int)
```
